```python
import utility
import os
import numpy as np
```

```python
# load sequence
sequence1 = np.loadtxt("9hong_mfcc.csv", delimiter=",")
# sequence1 = np.loadtxt("1hao_luo_mfcc.csv", delimiter=",")
# to be fixed
prefixes = ["1", "2", "3", "4", "5", "6", "7", "8", "9"]
sequence2_array = []
for p in prefixes:
    sequence2_array.append(np.loadtxt(f"./template_mfcc/{p}_mfcc.csv", delimiter=","))
print(sequence1.shape, len(sequence2_array))
```

```python
# dtw
cost_array = np.zeros(len(prefixes))
for p in prefixes:
    cost_array[int(p)-1] = utility.dtw_cost(sequence1, sequence2_array[int(p)-1])
best_prefix = np.argmin(cost_array) + 1
print(best_prefix)
```
